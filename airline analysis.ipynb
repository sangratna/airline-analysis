```python
tickets = pd.read_sql_query("""select * from tickets inner join bookings
                  on tickets.book_ref = bookings.book_ref""",conn)
tickets['book_date'] = pd.to_datetime(tickets['book_date'])
tickets['date'] = tickets['book_date'].dt.date
tickets
```

```python
x = tickets.groupby('date')[['date']].count()
plt.figure(figsize=(15,8))
plt.plot(x.index, x['date'],  linestyle='-', color='navy',marker='o',)
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Count of occurrences for each date')

# Display the plot
plt.show()
```

```python
query = """
    SELECT
        b.aircraft_code,
        b.flight_id,
        b.seat_counts,
        s.num_seats
    FROM
        (
            SELECT
                flights.aircraft_code,
                flights.flight_id,
                COUNT(*) AS seat_counts
            FROM
                boarding_passes
            INNER JOIN
                flights ON boarding_passes.flight_id = flights.flight_id
            GROUP BY
                flights.aircraft_code, flights.flight_id
        ) AS b
    inner JOIN
        (
            SELECT
                aircraft_code,
                COUNT(*) AS num_seats
            FROM
                seats
            GROUP BY
                aircraft_code
        ) AS s
    ON
        b.aircraft_code = s.aircraft_code;
"""

pd.read_sql_query(query, conn)

```
